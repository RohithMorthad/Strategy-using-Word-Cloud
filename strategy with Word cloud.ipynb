```python
import matplotlib.pyplot as plt
```

```python
from nltk.corpus import stopwords
from collections import Counter
```

```python
import nltk
nltk.download('stopwords')
tokens = word_tokenize(data)
tokens_n = [word for word in tokens
         if word not in stopwords.words('english')]
```

```python
count_freq = Counter(tokens_n)
most_freq = count_freq.most_common(50)
most_freq
```

```python
string = ""
for word, freq in most_freq:
    string = string + " " + word
```

```python
wrd_cld_model = WordCloud(width = 800, height = 800,background_color='pink', min_font_size= 4)
wordcloud = wrd_cld_model.generate(string)
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis=('off')
plt.tight_layout(pad = 0)
```
